```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
cancer
```

```python
dataset = pd.DataFrame(cancer.data, columns=cancer.feature_names)
dataset.head()
```

```python
dataset['target'] = cancer.target
dataset.head()
```

```python
dataset.to_csv('breast_cancer.csv', index=False)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(16, 9))# Target Counts
sns.countplot(x='target', data=dataset)
plt.show()
```

```python
dataset.corr()
```

```python
plt.figure(figsize=(25, 16))
sns.heatmap(dataset.corr(),cmap="cool", annot=True)
plt.show()
```

```python
X = dataset.drop(['target'], axis=1)
y = dataset['target']
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import accuracy_score#Machine Learning Model Training
```

```python
# XGBoost Classifier Without Scaled Data
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
y_pred_XGB = xgb.predict(X_test)
accuracy_score(y_test, y_pred_XGB)
```

```python
# With Scaled Data
xgb_scaled = XGBClassifier()
xgb_scaled.fit(X_train_scaled, y_train)
y_pred_XGB_Scaled = xgb_scaled.predict(X_test_scaled)
accuracy_score(y_test, y_pred_XGB_Scaled)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred_XGB)
plt.title('Confusion Matrix Heatmap')
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_XGB))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(xgb, X_train, y_train, cv=10)
scores, scores.mean()
```

```python
import pickle #saving Model
pickle.dump(xgb, open('model.pkl', 'wb'))
```

```python
#Loading the model
breast_cancer_detector = pickle.load(open('model.pkl', 'rb'))
```

```python
prediction = breast_cancer_detector.predict(X_test)
accuracy_score(y_test, prediction)
```
